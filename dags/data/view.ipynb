```python
# 1. Daily Price Movements
df['Daily_Range_M'] = df['High_M'] - df['Low_M']
df['Daily_Range_V'] = df['High_V'] - df['Low_V']
df['Price_Change_M'] = df['Close_M'] - df['Open_M']
df['Price_Change_V'] = df['Close_V'] - df['Open_V']
```

```python
# 2. Relative Changes (Percent Change)
df['Percent_Change_M'] = (df['Close_M'] - df['Open_M']) / df['Open_M'] * 100
df['Percent_Change_V'] = (df['Close_V'] - df['Open_V']) / df['Open_V'] * 100
```
